```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 2. LOADING THE DATASET

- The dataset was loaded for each course.
- The dataset was read to check if it has headers.

```python
data = pd.read_csv('/Users/joemande/Downloads/archive/googleplaystore.csv')
data.head(3)
```


### 3. DATA CLEANING

- To delete any blank cells

```python
data = data.dropna()
```

- To remove any duplicates

```python
data = data.drop_duplicates()
```

- To ensure that the data is consistent
  1. Check for data type
  2. Text Handling
  3. Change data type

```python
data.dtypes
```

```python
data['Installs']=data['Installs'].str.replace(',','')
data['Installs']=data['Installs'].str.replace('+','')
data['Price']=data['Price'].str.replace('$','')
data.head(3)
```

```python
data['Installs']=data['Installs'].astype('int')
data['Price']=data['Price'].astype('float')
data.dtypes
```

### 4. DATA ANALYSIS AND VISUALIZATION

- Total number of Installs for each genres
  1. Table
  2. Chart

```python
first_analysis = data.groupby(['Category'])['Installs'].sum()
pd.DataFrame(first_analysis)
```

```python
plt.subplots(figsize=(20,20))
plt.title('Total number of Installs for each genres')
plt.pie(first_analysis, radius=0.5, rotatelabels=60, labels=['ART_AND_DESIGN', 'AUTO_AND_VEHICLES', 'BEAUTY',
                                                 'BOOKS_AND_REFERENCE','BUSINESS','COMICS','COMMUNICATION',
                                                 'DATING', 'EDUCATION', 'ENTERTAINMENT','EVENTS','FAMILY','FINANCE',
                                                 'FOOD_AND_DRINK','GAME','HEALTH_AND_FITNESS','HOUSE_AND_HOME',
                                                 'LIBRARIES_AND_DEMO','LIFESTYLE','MAPS_AND_NAVIGATION',
                                                 'MEDICAL','NEWS_AND_MAGAZINES','PARENTING','PERSONALIZATION',
                                                 'PHOTOGRAPHY','PRODUCTIVITY','SHOPPING','SOCIAL','SPORTS','TOOLS',
                                                 'TRAVEL_AND_LOCAL','VIDEO_PLAYERS','WEATHER'])
plt.legend()
plt.show()
```

- Total number of App for each genres
  1. Table
  2. Chart

```python
second_analysis = data.groupby(['Category'])['App'].count()
pd.DataFrame(second_analysis)
```

```python
plt.subplots(figsize=(20,20))
plt.title('Total number of App for each genres')
plt.pie(second_analysis, radius=0.5, rotatelabels=60, labels=['ART_AND_DESIGN', 'AUTO_AND_VEHICLES', 'BEAUTY',
                                                 'BOOKS_AND_REFERENCE','BUSINESS','COMICS','COMMUNICATION',
                                                 'DATING', 'EDUCATION', 'ENTERTAINMENT','EVENTS','FAMILY','FINANCE',
                                                 'FOOD_AND_DRINK','GAME','HEALTH_AND_FITNESS','HOUSE_AND_HOME',
                                                 'LIBRARIES_AND_DEMO','LIFESTYLE','MAPS_AND_NAVIGATION',
                                                 'MEDICAL','NEWS_AND_MAGAZINES','PARENTING','PERSONALIZATION',
                                                 'PHOTOGRAPHY','PRODUCTIVITY','SHOPPING','SOCIAL','SPORTS','TOOLS',
                                                 'TRAVEL_AND_LOCAL','VIDEO_PLAYERS','WEATHER'])
plt.legend()
plt.show()
```

- Total number of App for each Content Rating
  1. Table
  2. Chart

```python
third_analysis = data.groupby(['Content Rating'])['App'].count()
pd.DataFrame(third_analysis)
```

```python
plt.subplots(figsize=(14,5))
plt.title('Total number of App for each Content Rating')
plt.barh(np.arange(len(third_analysis)), third_analysis, height=0.5, tick_label=['Adults only 18+','Everyone',
                                                                                 'Everyone 10+','Mature 17+',
                                                                                 'Teen','Unrated'], 
         color=['black', 'red', 'yellow', 'green', 'blue', 'violet'])
plt.ylabel('Content Rating')
plt.xlabel('App')
plt.show()
```

- Total number of App for each Type
  1. Table
  2. Chart

```python
fourth_analysis = data.groupby(['Type'])['App'].count()
pd.DataFrame(fourth_analysis)
```

```python
plt.subplots(figsize=(14,5))
plt.title('Total number of App for each Type')
plt.bar(range(len(fourth_analysis)), fourth_analysis, width=0.5, tick_label=['Free','Paid'], color=['blue', 'red'])
plt.ylabel('App')
plt.xlabel('Type')
plt.show()
```

- Total average price for each Content Rating
  1. Table
  2. Chart

```python
fifth_analysis = data.groupby(['Category'])['Price'].mean().round(2)
pd.DataFrame(fifth_analysis)
```

```python
plt.subplots(figsize=(14,12))
plt.title('Total average price for each Content Rating')
plt.barh(np.arange(len(fifth_analysis)), fifth_analysis, height=0.5, 
         tick_label=['ART_AND_DESIGN', 'AUTO_AND_VEHICLES', 'BEAUTY',
                                                 'BOOKS_AND_REFERENCE','BUSINESS','COMICS','COMMUNICATION',
                                                 'DATING', 'EDUCATION', 'ENTERTAINMENT','EVENTS','FAMILY','FINANCE',
                                                 'FOOD_AND_DRINK','GAME','HEALTH_AND_FITNESS','HOUSE_AND_HOME',
                                                 'LIBRARIES_AND_DEMO','LIFESTYLE','MAPS_AND_NAVIGATION',
                                                 'MEDICAL','NEWS_AND_MAGAZINES','PARENTING','PERSONALIZATION',
                                                 'PHOTOGRAPHY','PRODUCTIVITY','SHOPPING','SOCIAL','SPORTS','TOOLS',
                                                 'TRAVEL_AND_LOCAL','VIDEO_PLAYERS','WEATHER'])
plt.xlabel('Price')
plt.ylabel('Category')
plt.show()
```
